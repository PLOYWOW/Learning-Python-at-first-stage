```python
import pandas as pd

df = pd.read_csv("data2.csv")
df.fillna(0, inplace= True)
print(df.to_string())
```

```python
print(len(df))
```

```python
i = 0
for i in range(len(df)):
    if df.loc[i]["Duration"] < 45:
        df.loc[i, "Duration"] = 45
    else:
        if df.loc[i]["Duration"] > 60:
            df.loc[i, "Duration"] = 60
print(df.to_string())
```

```python
i = 0
for i in range(len(df)):
    if df.loc[i]["Pulse"] < 100:
        df.loc[i, "Pulse"] = 100
print(df.to_string())
```
